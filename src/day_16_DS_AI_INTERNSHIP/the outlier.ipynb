```python
import numpy as np
import pandas as pd

# Generate example dataset (you can replace this with your own dataset)
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Create DataFrame
df = pd.DataFrame({"values": data})

# 1️⃣ Calculate Mean and Standard Deviation
mean = df["values"].mean()
std_dev = df["values"].std()

print("Mean (μ):", mean)
print("Standard Deviation (σ):", std_dev)

# 2️⃣ Create Z-score column
df["z_score"] = (df["values"] - mean) / std_dev

# 3️⃣ Identify Outliers (|Z| > 3)
outliers = df[np.abs(df["z_score"]) > 3]

print("\nNumber of Outliers:", len(outliers))
print(outliers.head())
```
