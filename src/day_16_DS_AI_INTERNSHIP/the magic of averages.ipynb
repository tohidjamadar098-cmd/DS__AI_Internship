```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# 1️⃣ Create a heavily right-skewed dataset (Exponential distribution)
original_data = np.random.exponential(scale=2, size=10000)

# Plot original skewed data
plt.figure()
plt.hist(original_data, bins=30, density=True)
plt.title("Original Skewed Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# 2️⃣ Take 1000 samples of size 30 and compute their means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(original_data, size=30)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# 3️⃣ Plot histogram of the sample means
plt.figure()
plt.hist(sample_means, bins=30, density=True)
plt.title("Distribution of Sample Means (n=30, 1000 samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()

# Print comparison statistics
print("Original Data Mean:", np.mean(original_data))
print("Mean of Sample Means:", np.mean(sample_means))
print("Original Data Std Dev:", np.std(original_data))
print("Std Dev of Sample Means:", np.std(sample_means))
```
