```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "SquareFootage": [800, 1000, 1200, 1500, 1800, 2000, 2300, 2600],
    "Price": [150000, 180000, 210000, 260000, 300000, 340000, 390000, 450000],
    "City": ["Chicago", "Chicago", "Dallas", "Dallas", "New York", "New York", "Chicago", "Dallas"]
}

df = pd.DataFrame(data)

# -----------------------------
# 1 Scatter Plot (Numerical vs Numerical)
# -----------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="SquareFootage", y="Price", data=df)
plt.title("House Size vs Price")
plt.show()

# -----------------------------
# 2 Boxplot (Categorical vs Numerical)
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="City", y="Price", data=df)
plt.title("Price Distribution by City")
plt.show()
```
