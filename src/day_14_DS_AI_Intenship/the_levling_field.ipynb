```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Sample dataset
data = {
    "Age": [22, 25, 30, 35, 40, 45, 50],
    "Salary": [25000, 30000, 45000, 50000, 65000, 80000, 120000]
}

df = pd.DataFrame(data)

# -----------------------------
# 1️⃣ Standardization (Mean=0, Std=1)
# -----------------------------
scaler_std = StandardScaler()
df_standardized = pd.DataFrame(
    scaler_std.fit_transform(df),
    columns=df.columns
)

# -----------------------------
# 2️⃣ Normalization (0 to 1 range)
# -----------------------------
scaler_mm = MinMaxScaler()
df_normalized = pd.DataFrame(
    scaler_mm.fit_transform(df),
    columns=df.columns
)

print("Original Data:\n", df)
print("\nStandardized Data:\n", df_standardized)
print("\nNormalized Data:\n", df_normalized)

# -----------------------------
# 3️⃣ Histogram comparison (Salary)
# -----------------------------
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(df["Salary"])
plt.title("Original Salary")

plt.subplot(1,3,2)
plt.hist(df_standardized["Salary"])
plt.title("Standardized Salary")

plt.subplot(1,3,3)
plt.hist(df_normalized["Salary"])
plt.title("Normalized Salary")

plt.tight_layout()
plt.show()
```
