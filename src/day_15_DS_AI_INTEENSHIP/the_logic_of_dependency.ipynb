```python
import random

print("----- INDEPENDENT EVENTS -----")

# Analytical calculation
P_heads = 1/2
P_six = 1/6
P_independent = P_heads * P_six

print("Analytical Probability (Heads AND 6):", P_independent)

# Simulation
trials = 10000
independent_success = 0

for _ in range(trials):
    coin = random.choice(["H", "T"])
    die = random.randint(1, 6)
    
    if coin == "H" and die == 6:
        independent_success += 1

experimental_independent = independent_success / trials
print("Experimental Probability (Heads AND 6):", experimental_independent)


print("\n----- DEPENDENT EVENTS -----")

# Analytical calculation
P_first_red = 5/10
P_second_red_given_first = 4/9
P_dependent = P_first_red * P_second_red_given_first

print("Analytical Probability (Two Reds without replacement):", P_dependent)

# Simulation
dependent_success = 0

for _ in range(trials):
    bag = ["R"] * 5 + ["B"] * 5
    
    first = random.choice(bag)
    bag.remove(first)  # Without replacement
    
    second = random.choice(bag)
    
    if first == "R" and second == "R":
        dependent_success += 1

experimental_dependent = dependent_success / trials
print("Experimental Probability (Two Reds):", experimental_dependent)
```
